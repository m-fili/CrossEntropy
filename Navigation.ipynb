```python
import gym
from policy_based_cross_entropy.agent import Agent
from policy_based_cross_entropy.model import MLP
import matplotlib.pyplot as plt
from policy_based_cross_entropy.cross_entropy import CrossEntropy
from policy_based_cross_entropy.run import policy_based_run
```

### 1. Create Environment

```python
env = gym.make('MountainCarContinuous-v0')
```

### 2. Instantiate a model

```python
model = MLP(n_input=2, n_output=1, n_hidden=[16], random_state=110)
```

### 3. Instantiate an agent

```python
agent = Agent(env, model)
```

### 4. Define optimizer

```python
opt = CrossEntropy(n_params=model.tot_params,
                   pop_size=50,
                   elite_frac=0.1,
                   noise_scale=0.5)
```

### 5. Run

```python
scores = policy_based_run(agent, optimizer=opt, n_iterations=500, t_max=5000, gamma=1.0)
```

### 6. Learning Process

```python
plt.plot(scores)
plt.xlabel('episode')
plt.ylabel('score')
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```
